# Sierpinski

```python
vertexes = [[0, 0], [10, 0], [5, ((10 ** 2) - (5 ** 2)) ** 0.5]]
pts = [[5, 5]]
```

```python
from random import choice

for i in range(100_000):
    v = choice(vertexes)
    new_pt = [(pts[-1][0] + v[0]) / 2, (pts[-1][1] + v[1]) / 2]
    pts.append(new_pt)
```

```python
import numpy as np

pts = np.array(pts)
```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(8, 8))
ax.scatter(pts[:, 0], pts[:, 1], s=0.5);
```
